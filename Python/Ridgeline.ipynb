```python
import pandas as pd
df = pd.read_csv("../datasets/raw_data/temps_ridge.csv")
```

```python
regions = ['Alabama', 'Arizona', 'Arkansas', 'California', 'Colorado',
                 'Connecticut', 'Delaware', 'Florida', 'Georgia', 'Idaho',
                 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana',
                 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota',
                 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada',
                 'New Hampshire', 'New Jersey', 'New Mexico', 'New York',
                 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon',
                 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota',
                 'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington',
                 'West Virginia', 'Wisconsin', 'Wyoming', 'Alaska', 'Northeast',
                 'East North Central Region', 'Central Region', 'Southeast Region',
                 'West North Central Region', 'South Region', 'Southwest Region',
                 'Northwest Region', 'West Region', 'National']

dict_minimum = {estado: None for estado in regions}
dict_maximum = {estado: None for estado in regions}
    
```

## Minimum

```python
df1 = df[['Type', 'RegionName', 'Year', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']]
df1 = df1[(df1['Type']=="Minimum")]
df1 = df1.rename(columns={'Year': 'Month'})
df2 = df1.copy().T

df3 = df2[2:15]
df3.columns = df3.iloc[0]
df3 = df3[1:]
for i in range(59):
    state = df2.iloc[1,12*i-1]
    df_state = df3.copy()
    df_state = df_state.iloc[:, 12*i:12*i+12]
    dict_minimum[state] = df_state.copy()

dict_minimum["Alabama"]
```

```python
empty_dataframe = pd.DataFrame()
minimum_merged = pd.concat([empty_dataframe] + list(dict_minimum.values()), ignore_index=True)
minimum_merged
```

```python
minimum_merged.to_csv("../datasets/temps_ridgeline1.csv", index=False)
```
